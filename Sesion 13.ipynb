# Teorema del limite central

La suma de n variables aleatorias independientes, con un n > 30 tiende a una distribuci√≥n normal o la curva de campana incluso si las variables aleatorias originales no se distribuyen como una normal

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy.stats import expon, uniform
```

```python
poblacion = pd.DataFrame()
poblacion['numbers'] = expon.rvs(40, size=100000)
```

```python
poblacion['numbers'].hist(bins=100)
```

```python
muestra_promedio = []

tamano = 5000

for i in range(tamano):
    muestra_promedio.append(poblacion.sample(n=100).mean().values[0])
```

```python
fig, ax = plt.subplots()
ax.hist(muestra_promedio, bins=50, alpha=0.5)
ax.set(xlabel='Valor promedio', ylabel='Frecuencia')
ax.grid()
```
