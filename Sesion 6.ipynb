```python
import scipy.stats
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
#se pone inline para ejecutar dentro del jupyter
```

```python
df = pd.read_csv('bicicletas-compartidas.csv')
```

# Variables Categoricas

```python
df.columns
```

```python
y = df['cuartil-ano'].apply(lambda x: 'Cuartil' + str(int(x)))
```

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
```

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuartiles del año')
ax.set_ylabel('Frecuencia')
```

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuartiles del año')
ax.set_ylabel('Frecuencia')
ax.patches[2].set_facecolor('red')
```

```python
fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels = y.value_counts().index)
```

```python
fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels = y.value_counts().index)
ax.set_title('Diagrama de pie')
```

# Variables Numericas

```python
y = df['viento']
```

```python
fig, ax = plt.subplots()
ax.hist(y, bins=30)
```

```python
fig, ax = plt.subplots()
ax.hist(y, bins=30)
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuencia')
plt.axvline(np.mean(y), c='r', linestyle='--', label='Promedio')
```

```python
fig, ax = plt.subplots()
ax.hist(y, bins=30)
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuencia')
plt.axvline(np.mean(y), c='r', linestyle='--', label='Promedio')
plt.axvline(np.mean(y)+np.std(y), c='k', linestyle='--', label='+1-desviacion')
plt.axvline(np.mean(y)-np.std(y), c='k', linestyle='--', label='+1-desviacion')
ax.legend()
```

```python
y = df['bicis-compartidas']
fig, ax = plt.subplots()
ax.boxplot(x = y)
```

```python
y = df['bicis-compartidas']
fig, ax = plt.subplots()
sns.boxplot(x = 'cuartil-ano', y = 'bicis-compartidas', data=df)
```

```python
fig, ax = plt.subplots()
ax.scatter(df['viento'], df['bicis-compartidas'])
```

```python
fig, ax = plt.subplots()
ax.scatter(df['viento'], df['bicis-compartidas'], alpha=0.03)
ax.set_title('Distribucion conjunto de viento y bicis-compartidas')
ax.set_xlabel('Viento')
ax.set_ylabel('Bicis-compartidas')
```
